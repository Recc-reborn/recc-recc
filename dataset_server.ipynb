```python
def index_clear_names(list_names: list) -> list:
    list_caracters = ["/", ";"]
    result = []
    for i in range(len(list_names)):
        for char in list_caracters:
            if (char in list_names[i]):
                result.append(i)
                break
    
    return result
```

##### Create a function that will check if the artist name already exist on the DB

```python
from sqlalchemy import Engine, text


def already_exist(db_engine: Engine, artist_name: str):
    try:
        with db_engine.connect() as conn:
            result = conn.execute(text("select id from artists where name=:name", {"name": artist_name}))
            array_of_results = [dict(dict_row) for dict_row in result.mappings()]
            if(len(array_of_results)):
                return True
            return False
    except:
        print(f"This name raised an exception: {artist_name}")
        return False
```

```python
from db.create_db_engine import create_db_engine
engine = create_db_engine()
```

```python
already_exist(engine, "2")
```

```python
def get_missing_artist(artists_names):
    missing_artist = []
    for artist in artists_names:
        if (not already_exist(engine, artist)):
            missing_artist.append(artist)
    
    return missing_artist
```
